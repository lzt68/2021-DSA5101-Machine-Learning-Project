# Export value counts figure

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
for col in ['job', 'marital', 'education', 'default', 'housing', 'loan', 'contact', 'poutcome', 'y']:
    plt.figure(figsize = (4, 5))
    sns.countplot(x = col, data = data, order=data[col].value_counts().index)
#     sns.barplot(x = data[col].value_counts().index, y = data[col].value_counts())
    if col == 'job':
        plt.xticks(rotation = 90)
    plt.savefig(".\\figure\\" + col +"_value_counts.png", bbox_inches='tight')
```
